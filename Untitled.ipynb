```python
import os
import librosa 
import numpy as np

# List of folders containing audio files
audio_folders = ["C:/Users/rodea/Documents/GitHub/Infant-Crying-Classification/Data/belly_pain",
                 "C:/Users/rodea/Documents/GitHub/Infant-Crying-Classification/Data/burping"
                "C:/Users/rodea/Documents/GitHub/Infant-Crying-Classification/Data/discomfort",
                 "C:/Users/rodea/Documents/GitHub/Infant-Crying-Classification/Data/hungry"
                "C:/Users/rodea/Documents/GitHub/Infant-Crying-Classification/Data/tired"]

# An empty list to store processed audio data
train_audio_data = []

# Parameters for audio processing
sr = 44100  # Sample rate
duration = 5  # Duration of each audio clip in seconds

# Iterate through each folder
for folder in audio_folders:
    # List all audio files in the folder
    files = os.listdir(folder)

    # Iterate through each audio file
    for file in files:
        # Get the full file path
        file_path = os.path.join(folder, file)

        # Load the audio file
        audio, _ = librosa.load(file_path, sr=sr, duration=duration)

        # If audio is shorter than the desired duration, pad it with zeros
        if len(audio) < sr * duration:
            padding = sr * duration - len(audio)
            audio = np.pad(audio, (0, padding), mode='constant')

        # Add processed audio to the list
        train_audio_data.append(audio)

# Convert the list of audio data into a numpy array
concatenated_audio_data = np.array(train_audio_data)

print("Shape of concatenated audio data:", concatenated_audio_data.shape)
```
